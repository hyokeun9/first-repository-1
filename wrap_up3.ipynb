시가총액을 기반으로 다시 계산

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

# 1️⃣ 시가총액 기반 일별 가중 수익률 계산 (예: 현대차, 기아)
total_market_cap = hyundai['Market Capitalization'] + kia['Market Capitalization']
weighted_return = (hyundai['return'] * hyundai['Market Capitalization'] +
                   kia['return'] * kia['Market Capitalization']) / total_market_cap

# 2️⃣ 회귀 분석: 대장주 -> KRX 자동차 섹터
X = pd.DataFrame({
    'Hyundai': hyundai['return'],
    'Kia': kia['return']
})
y = krx_auto['return']

model = LinearRegression()
model.fit(X, y)

print("Regression Results (Daily Returns):")
print(f"Hyundai contribution: {model.coef_[0]:.3f}")
print(f"Kia contribution: {model.coef_[1]:.3f}")
print(f"Intercept: {model.intercept_:.3f}")
print(f"R^2: {model.score(X, y):.3f}")

# 3️⃣ 이동 상관 (Rolling Correlation) 계산
window = 20  # 20일 이동 상관
rolling_corr_hy = hyundai['return'].rolling(window).corr(krx_auto['return'])
rolling_corr_kia = kia['return'].rolling(window).corr(krx_auto['return'])

plt.figure(figsize=(12,5))
plt.plot(rolling_corr_hy, label='Hyundai vs KRX Auto', linewidth=2)
plt.plot(rolling_corr_kia, label='Kia vs KRX Auto', linewidth=2)
plt.title(f'{window}-Day Rolling Correlation: Hyundai & Kia vs KRX Auto')
plt.xlabel('Date')
plt.ylabel('Correlation')
plt.legend()
plt.grid(True)
plt.show()

# 4️⃣ 누적 수익률 시각화 (기준일 100)
cum_hy = (1 + hyundai['return']).cumprod() * 100
cum_kia = (1 + kia['return']).cumprod() * 100
cum_auto = (1 + krx_auto['return']).cumprod() * 100

plt.figure(figsize=(12,6))
plt.plot(hyundai['date'], cum_hy, label='Hyundai', linewidth=2)
plt.plot(kia['date'], cum_kia, label='Kia', linewidth=2)
plt.plot(krx_auto['date'], cum_auto, label='KRX Auto', linewidth=2)
plt.title('Cumulative Returns: Hyundai, Kia vs KRX Auto')
plt.xlabel('Date')
plt.ylabel('Cumulative Return (Base=100)')
plt.legend()
plt.grid(True)
plt.show()
```

- 두 대장주만으로는 KRX 자동차 섹터의 변동성을 충분히 설명하기 어려움

- 현실적으로는 전체 섹터 종목 + 가중치 + 기간별 변동성 반영이 필요

```python
#단순하게 오를확률

import pandas as pd

def rising_probability(df, name, start='2023-11-01', end='2025-10-31'):
    df = df.copy()
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)

    df1 = df.loc[start:end]
    df1['ret'] = df1['Closing Price'].pct_change() * 100
    r = df1['ret'].dropna()

    if len(r) == 0:
        print(f" {name}: 해당 기간에 데이터가 없습니다.")
        return {'Name': name, 'Rising Probability': None}

    positive = (r > 0).sum()
    total = len(r)
    positive_rate = positive / total

    return {
        'Name': name,
        'Positive Days': positive,
        'Total Days': total,
        'Rising Probability': round(positive_rate * 100, 2)
    }

results = []
results.append(rising_probability(kospi, 'KOSPI'))
results.append(rising_probability(hyundai, 'Hyundai Motor'))
results.append(rising_probability(kia, 'Kia'))
results.append(rising_probability(krx_auto, 'KRX Automobiles'))

summary = pd.DataFrame(results)
print(summary)
```
