```python
df = fdr.DataReader('005930', '2010-01-01')
```

```python
df.head
```

```python
df2 = pd.DataFrame()
df1 = df.loc['2022-01-01':'2022-12-31']
df2['ret'] = (df1['Close'].pct_change())*100
df3 = df2['ret']
r = df3.dropna()
len
```

```python
positive = 0
n = len(r)
for i in range(n):
  if r[i] > 0:
    positive += 1
positive
```

```python
negative = n - positive
```

```python
negative
```

```python
# 주가가 오를 확률
```

```python
positive_rate = positive / n
```

```python
positive_rate
```

```python
#조합공식
#예시 ) 5일 간의 데이터 중 3일을 선택해 평균 주가 변동 구하기
```

```python
# 이항분포
```

```python
# 5일 중에 3일동안 주가가 오를 확률
import math
a = math.factorial(5)
b = math.factorial(3) - math.factorial(2)
number = a/b

c = ((positive_rate)**3)*((1-positive_rate)**2)

prob = number * c
prob

```
